```python
import os
import matplotlib.pyplot as plt
```

```python
path = '/gdrive/MyDrive/Universidad/Tesis/Datasets/dataset/data'
```

```python
di = {'test':{"Tronco":0,"1 año":0,"2 años":0,"3 años o más":0},'train':{"Tronco":0,"1 año":0,"2 años":0,"3 años o más":0},'val':{"Tronco":0,"1 año":0,"2 años":0,"3 años o más":0}}
```

```python
labels = ["Tronco","1 año","2 años","3 años o más"]
```

```python
for i in os.listdir(path):
  for k in os.listdir(path+'/'+i+'/labels/'):
    with open(path+'/'+i+'/labels/'+k,'r') as t:
      lines = t.readlines()
      for ann in lines:
        di[i][labels[int(ann[0])]] = di[i][labels[int(ann[0])]]+1
```

```python
di
```

```python
width = 0.8
colors = ['#DE7C7D', '#CC2B52','#AF1740','#740938']
plt.subplot(1, 3, 1)
plt.bar(di["train"].keys(), di["train"].values(), width, color=colors)
plt.title("Conjunto de entrenamiento", fontsize=10)
plt.xticks(rotation=60)
for i, v in enumerate(di["train"].values()):
    plt.text( i,v + 3, str(v), color='black', horizontalalignment='center',fontsize=8)

plt.subplot(1, 3, 2)
plt.bar(di["val"].keys(), di["val"].values(), width, color=colors)
plt.title("Conjunto de validacion", fontsize=10)
plt.xticks(rotation=60)
for i, v in enumerate(di["val"].values()):
    plt.text( i,v + 1, str(v), color='black', horizontalalignment='center',fontsize=8)

plt.subplot(1, 3, 3)
plt.bar(di["test"].keys(), di["test"].values(), width, color=colors)
plt.title("Conjunto de prueba", fontsize=10)
plt.xticks(rotation=60)
for i, v in enumerate(di["test"].values()):
    plt.text( i,v + 1, str(v), color='black', horizontalalignment='center',fontsize=8)

plt.tight_layout()
plt.suptitle('Instancias de cada clase')
plt.subplots_adjust(top=0.88)
```
